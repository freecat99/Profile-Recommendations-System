```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("SMUsers.csv")
```

```python
df
```

```python
df['DOB'] = pd.to_datetime(df['DOB'])
today = pd.to_datetime('today')
df['Age'] = round((today-df['DOB']).dt.days/365).astype(int)
```

```python
df['Age']
```

```python
df
```

```python
df.drop(df[df['Country'] != 'India'].index, inplace = True)
```

```python
df.drop(columns=['DOB'], inplace=True)
```

```python
df
```

```python
df['Interests'] = df['Interests'].str.split(", ").apply(lambda x: [i.strip("'") for i in x])
```

```python
df
```

```python
df['Interests'] = df['Interests'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
df
```

```python
df['Age'] = df['Age'].astype(str)
```

```python
df
```

```python
df['Tags'] = df['Gender'] + " "+df['City'] + " " +df['Age'] + " "+ df['Interests'].apply(lambda x: ' '.join(x)) + " "
```

```python
df
```

```python
final_df = df[['UserID', 'Name', 'Tags']]
```

```python
final_df = final_df.reset_index(drop=True)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
cv
```

```python
cv.fit_transform(final_df['Tags']).toarray()
```

```python
vectors = cv.fit_transform(final_df['Tags']).toarray()
```

```python
vectors
```

```python
len(cv.get_feature_names_out())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])), reverse=True, key = lambda x:x[1])[1:6]
```

```python
def recommend(profile):
    person_index = final_df[final_df['Name'] == profile].index[0]
    distances = similarity[person_index]
    recc_list = sorted(list(enumerate(distances)), reverse=True, key = lambda x:x[1])[1:6]

    for i in recc_list:
        user_id = int(final_df.iloc[i[0]]['UserID'])
        user_info = final_df[final_df['UserID'] == user_id][['UserID', 'Name', 'Tags']]
        print(user_info)

    return 
```

```python
recommend("George Macon")
```
